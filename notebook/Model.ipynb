### Ouliers

```python
df_outliers = df_clean.copy()
```

```python
num = df_clean.select_dtypes(exclude='object').columns
```

```python
# Melakukan identifikasi kolom yang mengalami outliers
print(f'Jumlah baris: {len(df_outliers)}')

outlier = []
no_outlier = []
is_outlier = []
low_lim = []
high_lim = []


filtered_entries = np.array([True] * len(df_outliers))
for col in num:
    Q1 = df_outliers[col].quantile(0.25)
    Q3 = df_outliers[col].quantile(0.75)
    IQR = Q3 - Q1
    low_limit = Q1 - (IQR * 1.5)
    high_limit = Q3 + (IQR * 1.5)

    #mulai filter outlier
    filter_outlier = ((df_outliers[col] >= low_limit) & (df_outliers[col] <= high_limit))
    outlier.append(len(df_outliers[~filter_outlier]))
    no_outlier.append(len(df_outliers[filter_outlier]))
    is_outlier.append(df_outliers[col][~filter_outlier].any())
    low_lim.append(low_limit)
    high_lim.append(high_limit)

    filtered_entries = ((df_outliers[col] >= low_limit) & (df_outliers[col] <= high_limit)) & filtered_entries

print("Outlier All Data :", len(df_outliers[~filtered_entries]))
print("Not Outlier All Data :", len(df_outliers[filtered_entries]))
print()

pd.DataFrame({
    "Column Name":num,
    "is Outlier": is_outlier,
    "Lower Limit": low_lim,
    "Upper Limit": high_lim,
    "Outlier":outlier,
    "No Outlier":no_outlier
})
```

**Observasi:**  
Terdapat banyak data numerik yang mengalami outlier sehingga perlu dilakukan handle Outliers, saya menggunakan metode IQR utnuk mengatasi hal tersebut.

```python
# Melakukan handle outlier dengan menggunakan metode IQR
columns_is_outlier = df_outliers[num].drop(['loan_amnt', 'loan_target'], axis=1)

print('Jumlah baris sebelum handle outlier :', df_outliers.shape[0])
# Iterasi melalui setiap kolom untuk menangani outlier
for column in columns_is_outlier:
    Q1 = df_outliers[column].quantile(0.25)
    Q3 = df_outliers[column].quantile(0.75)
    IQR = Q3 - Q1

    # Menghitung batas bawah dan atas outlier
    lower_limit = Q1 - (1.5 * IQR)
    upper_limit = Q3 + (1.5 * IQR)

    # Menghapus outlier dari DataFrame
    df_outliers = df_outliers[(df_outliers[column] >= lower_limit) & (df_outliers[column] <= upper_limit)]

print('Jumlah baris setelah handle outlier :', df_outliers.shape[0])
```

# Modeling

Beberapa metode yang akan dibandingan dan di evaluasi pada proyek ini:
1. Logistic Regression
2. Random Forest
3. Decision Tree
4. XGBoost
5. Naive Bayes
6. SVM

Kemudian, Metriks evaluasi yang akan digunakan pada proyek ini:
1. Accuracy
2. Precision
3. Recall
4. F1-Score
5. ROC AUC

```python
# Selection Model
models = {}

# Logistic Regression
models['Logistic Regression'] = LogisticRegression()

# Decision Trees
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
models['Random Forest'] = RandomForestClassifier()

# XGBoost
models['XGBoost'] = XGBClassifier()

# Naive Bayes
models['Naive Bayes'] = GaussianNB()

# SVM
models['Support Vector Machines'] = LinearSVC()
```

```python
accuracy, precision, recall, f1, roc_auc = {}, {}, {}, {}, {}

for key in models.keys():
    # Melakukan train setiap model
    models[key].fit(X_train, y_train)

    # Melakukan prediksi model
    predict = models[key].predict(X_test)

    # Menghitung metrik Accuracy, Precision, Recall, F1 Score, dan ROC AUC
    accuracy[key] = accuracy_score(predict, y_test)
    precision[key] = precision_score(predict, y_test)
    recall[key] = recall_score(predict, y_test)
    f1[key] = f1_score(predict, y_test)
    roc_auc[key] = roc_auc_score(predict, y_test)
```

```python
# Menampilkan metriks dari setiap model
df_model = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall', 'F1-Score', 'Roc_auc'])
df_model['Accuracy'] = accuracy.values()
df_model['Precision'] = precision.values()
df_model['Recall'] = recall.values()
df_model['F1-Score'] = f1.values()
df_model['Roc_auc'] = roc_auc.values()

round(df_model, 2)
```

**Observasi:**  
Berdasarkan hasil pemodelan yang telah dilakukan, diketahui bahwa model Random Forest dan XGBoost memiliki nilai metrik tertinggi dibandingkan dengan model lainnya. Sehingga, Evaluasi model dalam proyek ini akan difokuskan pada kedua model tersebut.

### Confusion Matriks

Diketahui:

- 0 = Negatif (Nasabah yang tidak berisiko gagal bayar pinjaman)

```python
# Confussion Matrix Random Forest
matrix_rf = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize=(3, 3))
sns.heatmap(matrix_rf, annot = True, fmt='d', cmap="Blues")
plt.title('Confussion Matrix Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### Cross Validation

```python
rf_scores = cross_val_score(rf, X_train, y_train, cv=10, scoring='f1')
xgb_scores = cross_val_score(xgb, X_train, y_train, cv=10, scoring='f1')
```

### Kurva ROC AUC

```python
# Predict Probabilities
y_probs_rf = rf.predict_proba(X_test)[:,1]
y_probs_xgb = xgb.predict_proba(X_test)[:,1]

# Hitung nilai AUC
rf_auc = roc_auc_score(y_test, y_probs_rf)
xgb_auc = roc_auc_score(y_test, y_probs_xgb)

print(f'Random Forest: AUROC = {rf_auc:.3f}')
print(f'XGBoost: AUROC = {xgb_auc:.3f}')
```

```python
# Hitung nilai TPR dan FPR
rf_fpr, rf_tpr, thresholds  = roc_curve(y_test, y_probs_rf)
xgb_fpr, xgb_tpr, thresholds  = roc_curve(y_test, y_probs_xgb)

# Plot ROC curve
plt.figure(figsize=(5, 5))
plt.plot(rf_fpr, rf_tpr, color='darkorange', lw=2, label='Random Forest (AUROC = %0.3f)' % rf_auc)
plt.plot(xgb_fpr, xgb_tpr, color='darkorange', lw=2, label='XGBoost (AUROC = %0.3f)' % xgb_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

plt.title(f'ROC Plot')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")

plt.show()
```
